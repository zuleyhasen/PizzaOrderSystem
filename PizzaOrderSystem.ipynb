```python
import csv
import datetime

#csv, datetime import ettik.



#Pizza üst sınıfı ve get_description, get_cost metodları

class Pizza:
    def __init__(self):
        self.description = 'Unknown Pizza'
    def get_description(self):
        return self.description
    def get_cost(self):
        return self.cost
    
#Pizza üst sınıfının alt sınıfı olarak pizza türleri ve üst sınıftaki methodları kullandığımız yer
class ClassicPizza(Pizza):
        def __init__(self):
            self.description = 'Klasik Pizza (Domates Sosu,Mozarella Peyniri, Jambon, Sosis, Zeytin, Mısır, Mantar) '
            print(self.description)
            self.cost = 45
class MargaritaPizza(Pizza):
        def __init__(self):
            self.description = 'Margarita Pizza (Domates Sosu,Mozarella Peyniri, Domates, Fesleğen, Zeytinyağı) '
            print(self.description)
            self.cost = 35
class TurkPizza(Pizza):
        def __init__(self):
            self.description = 'Turk Pizza (Domates Sosu,Mozarella Peyniri, Sucuk, Kavurma, Keçi Peyniri, Soğan)'
            print(self.description)
            self.cost = 60
class PlainPizza(Pizza):
        def __init__(self):
            self.description = 'Sade Pizza (Domates Sosu,Mozarella Peyniri)'
            print(self.description)
            self.cost = 20
    
        
        
#Soslar için oluşturulan Decorator üst sınıfı ve metodları
 
class Decorator:
    def __init__(self, pizza):
        self.pizza = pizza
    def get_description(self):
      return self.pizza.get_description() + ", " + self.description
    def get_cost(self):
      return self.pizza.get_cost() + self.cost

#Decorator üst sınıfının alt sınıfı olarak malzeme türleri 
class Olives(Decorator):
    def __init__(self, pizza):
            self.pizza = pizza
            self.description = 'Zeytin'
            self.cost = 5
            def get_description(self):
                return self.pizza.get_description() + ', ' + self.description
            def get_cost(self):
                return self.pizza.get_cost() + self.cost

class Mushrooms(Decorator):
    def __init__(self, pizza):
            self.pizza = pizza
            self.description = 'Mantar'
            self.cost = 5
            def get_description(self):
                return self.pizza.get_description() + ', ' + self.description
            def get_cost(self):
                return self.pizza.get_cost() + self.cost
            
class GoatCheese(Decorator):
    def __init__(self, pizza):
            self.pizza = pizza
            self.description = 'Keçi Peyniri'
            self.cost = 10
            def get_description(self):
                return self.pizza.get_description() + ', ' + self.description
            def get_cost(self):
                return self.pizza.get_cost() + self.cost
                
class Meat(Decorator):
    def __init__(self, pizza):
            self.pizza = pizza
            self.description = 'Et'
            self.cost = 15
            def get_description(self):
                return self.pizza.get_description() + ', ' + self.description
            def get_cost(self):
                return self.pizza.get_cost() + self.cost
            

class Onions(Decorator):
    def __init__(self, pizza):
            self.pizza = pizza
            self.description = 'Sogan'
            self.cost = 5
            def get_description(self):
                return self.pizza.get_description() + ', ' + self.description
            def get_cost(self):
                return self.pizza.get_cost() + self.cost

class Corn(Decorator):
    def __init__(self, pizza):
            self.pizza = pizza
            self.description = 'Mısır'
            self.cost = 5
            def get_description(self):
                return self.pizza.get_description() + ', ' + self.description
            def get_cost(self):
                return self.pizza.get_cost() + self.cost
            
            

#main fonksiyonu
def main():

  sum = 0
  pizza_descriptions = ' '
  
  print("*" * 50)
  print("          *  PİZZA SİPARİŞ SİSTEMİ  *")
  print("*" * 50)
  print()
  print()
#menüyü menu.txt dosyası ile yazdırma
  with open("Menu.txt", "w") as file:
    file.write("* Lütfen Bir Pizza Tabanı Seçiniz:\n1: Klasik\n2: Margarita\n3: TürkPizza\n4: Sade Pizza\n* ve seçeceğiniz sos:\n11: Zeytin\n12: Mantarlar\n13: Keçi Peyniri\n14: Et\n15: Soğan\n16: Mısır\n* Teşekkür ederiz!")

  with open("Menu.txt", "r") as file:
    print(file.read())

#while döngüsü ile birden fazla sipariş verebilme seçeneği sunduk
  while True:
    print()
    print('Lütfen listede gördüğünüz pizzalardan birini seçiniz!!')
    print()
    pizza_choice = int(input('Pizza seçimi: ')) #pizza seçimi için input
    print()
    if pizza_choice == 1:
        pizza = ClassicPizza()
    elif pizza_choice == 2:
        pizza = MargaritaPizza()
    elif pizza_choice == 3:
        pizza = TurkPizza()
    elif pizza_choice == 4:
        pizza = PlainPizza()
    else:
        print('Geçersiz seçim!')
        #yanlış input girildiğinde tekrar pizza seçim ekranına dönmek için main fonksiyonunu tekrar çağırıyoruz
        main()
    
    print()
    print('Lütfen listede gördüğünüz soslardan birini seçiniz!')
    print()
    #Sos seçimi için input
    decorator_choice = int(input('Sos seçimi: '))
    
    if decorator_choice == 11:
        pizza = Olives(pizza)
    elif decorator_choice == 12:
        pizza = Mushrooms(pizza)
    elif decorator_choice == 13:
        pizza = GoatCheese(pizza)
    elif decorator_choice == 14:
        pizza = Meat(pizza)
    elif decorator_choice == 15:
        pizza = Onions(pizza)
    elif decorator_choice == 16:
        pizza = Corn(pizza)
    else:
        print('Geçersiz seçim!')
        #tekrar yanlış input için main fonksiyonunu tekrar çağırma
        main()
    #birden fazla seçilen pizzaların fiyatlarının toplamı
    sum += pizza.get_cost()
    pizza_descriptions += "  " + pizza.get_description()
    

    print()
    #Sipariş edilen pizzanın açıklaması ve fiyatı
    print('Pizza: ' + pizza.get_description())
    print('Ücret: ' + str(pizza.get_cost()))

    print()
    #Birden fazla sipariş seçeneği.Eğer input 2 ise while loopun biterek bi sonraki adıma geçilmesi
    print('1. Yeni sipariş')
    print('2. Ödeme')
    order_again = int(input("Seçiniz :"))

    if order_again == 2:
      break
    
  print()
  print("Toplam Ücret:"+ str(sum) +" Türk Lirası")
  print()
  print(input("Siparişi onaylıyorsanız lütfen Enter'a basınız!!"))

#Kullanıcı ve ödeme bilgilerini giriş ve database dosyasına kaydetme


  name = input('Lütfen isminizi giriniz: ')

  tc_no = input('Lütfen TC kimlik numaranızı giriniz: ')
  #eğer 11 haneden eksik veya fazlaysa"yanlış veya eksik girdiniz!" yazıp tekrar soruyor
  while len(str(tc_no)) != 11:
      print("Yanlış veya eksik girdiniz!")
      tc_no = input('Lütfen TC kimlik numaranızı giriniz: ')



  creditcard = input('Lütfen kredi kartı numaranızı giriniz: ')
  #eğer 16 haneden eksik veya fazlaysa "yanlış veya eksik girdiniz!" yazıp tekrar soruyor
  while len(creditcard) != 16:
      print("Yanlış veya eksik girdiniz!")
      creditcard = input('Lütfen kredi kartı numaranızı giriniz: ')
      
  
  password = input('Lütfen kredi kartı şifrenizi giriniz: ')
  #eğer 4 haneden eksik veya fazlaysa "yanlış veya eksik girdiniz!" yazıp tekrar soruyor  
  while len(password) != 4 :
      print("Yanlış veya eksik girdiniz!")
      password = input('Lütfen kredi kartı şifrenizi giriniz: ')

  order_time = datetime.datetime.now()
  with open('Orders_Database.csv', 'a') as file:
    file.write('kullanıcı adı, TC Kimlik no, Kart No, Sifre, Siparis, Siparis zamanı \n')
    writer = csv.writer(file)
    writer.writerow([name, tc_no, creditcard, password, pizza_descriptions, order_time])


  print()
  print("*" * 70)
  print('Siparişiniz Oluşturuldu !!')
  print('Pizza Sipariş Sistemimizi kullandığınız için teşekkürler :)')
  print("*" * 70)

    

if __name__ == "__main__":
  main()
```
